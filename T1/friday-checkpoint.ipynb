```python
import matplotlib.pyplot as plt
import networkx as nx

data={ "p1":[0,1,1,0,1],"p2":[0,0,1,0,1], "p3":[1,1,0,1,1], "p4":[1,1,1,0,0],"p5":[1,1,0,0,0]}


g=nx.DiGraph()
g.add_nodes_from(data.keys())

for i,j in enumerate(data):
    for k,h in enumerate(data[j]):
        if h==1:
            g.add_edge(j,list(data.keys())[k])

nx.draw(g,with_labels=True , node_size=1000 ,node_color="red",width=5)
plt.show()
```
